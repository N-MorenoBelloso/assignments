## Graphs

To make the graphs we need to make selections of the data. Please use "dplyr" (commands "select" and "filter") and the "pipe". The first assignment is to make a scatterplot using "ggplot". The graph contains data for all countries for the year 2014. Therefore make a dataframe with dplyr first, called "graph1" with the required data.
Then, make a scatterplot with "Cost" on the x-axis and "LE" on the y-axis. You have to add a regression line with a shaded area for the confidence interval.

```r
graph1 <- oecd %>% 
        select(Country, Year, Cost, LE) %>%
        filter(Year == "2014")        # filtering the data to only be left with the data for the year 2014

ggplot(graph1, aes(x=Cost, y=LE)) +        # creating the graph with the correct x- and y-axis
        geom_point(shape=1) +       # choosing hollow circles
        geom_smooth(method=lm) +       # adding a regression line which automatically includes a 95% confidence region.
        labs(title="Scatter plot of Cost and LE") # adding a title to the graph
```

The second assignment is to make a line graph. You will use the years 2005-2015 for 5 countries (you can choose your own countries). Make a new dataframe called graph2 with the required data. Then, make a line graph with "Year"  and "Cost" on the y-axis. There should be coloured lines per country and a legend attached. 

```r
graph2 <- oecd %>%
        select(Country, Year, Cost, LE) %>%        # selecting the relevant columns of the dataframe oecd
        filter(Country %in% c("Greece", "Portugal", "Spain", "Italy", "Turkey")) # filtering 5 countries

ggplot(graph2, aes(Year, Cost, color=Country)) + # choosing the x- and y-axis and color by country
    geom_line() +        # making the ggplot into a line graph
    labs(title="Cost in 5 OECD countries from 2005 to 2015")   # adding a title to the graph
```

## Efficiency scores

The last assignment is to compute simple effiency scores. We will use the year 2014 for all countries (i.e. the dataframe "graph1").
Make a new dateframe called "eff" (using graph1). Please use the "pipe" to execute to following steps:
Select all the variables except the NA's (hint: use na.omit) and create a new column called "Ratio" with mutate. Ratio is defined as "Cost/LE" (cost per output).
Create a new column called "Score" with mutate. Score is the ratio per country divided by the lowest ratio in the dataframe.
To complete this dataframe order the dataset using "arrange" based on score (i.e. lowest Score first).


Finally make a bar chart using ggplot. Countries should be on the x-axis and "Score" on the y-axis. 



You can earn extra points if you are able to order the graph based on scores (lowest first, highest last), if you are able to make the x-axis readable, make the bar of the Netherlands red and the other bars blue and/or change the name of the y-axis to "Efficiency scores according to [your name]" 


The figure should like this:

![My Figure](eff.png)

```r
eff <- graph1 %>%
        na.omit %>%        # omitting the NA's
        mutate(Ratio = Cost / LE) %>%        # creating a new column called "Ratio"
        mutate(Score = min(Ratio) / Ratio) %>%        # creating a new column called "Score". Score is the lowest ratio in the dataframe divided by the ratio per country
        arrange(desc(Score))        # arranging the the column "Score" in descending order, to make sure the graph will resemble the example graph

eff$fill <- ifelse((eff$Country == "Netherlands"), "red", "blue")        # creating an additional column called "fill", which attaches the value "red" to the Netherlands and "blue" to everything else

ggplot(eff, aes(reorder(Country, Score), Score, fill=fill)) +        # creating the ggplot. Note that I choose to color by variable, in particular the variable "fill" I created above.
        geom_bar(stat="identity") +        # specifying that it must be a barchart
        theme(axis.text.x = element_text(angle = 60, hjust = 1)) +        # to make the x-axis readable, the labels are rotated at a 60 degree angle
        labs(x="",y="Efficiency scores according to Natalia")+        # adding the required y-axis name
        scale_fill_identity() +       # needed to make sure the Netherlands is colored red and the rest is colored blue, in accordance with what the column I created above and the fill option I choose.
        labs(title="Efficiency scores of OECD countries") # adding a title to the graph
```
